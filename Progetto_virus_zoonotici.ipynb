# Cambiamenti climatici ed epidemie: nuove sfide per l'umanità
     
     
   *Il cambiamento climatico rappresenta la più grande minaccia del XXI secolo per la salute globale 
   (Organizzazione Mondiale della Sanità, 2003)*



I cambiamenti climatici rappresentano una delle maggiori sfide del XXI secolo e il loro impatto sulla Terra è una delle priorità per la tutela della salute pubblica; gli effetti, infatti, di detti cambiamenti, sempre più violenti ed estremi, minacciano la società contemporanea, sotto ogni aspetto e in tutti i suoi ambiti. I rischi sanitari, potenzialmente associati al clima che cambia, devono assolutamente ricevere maggiore attenzione da parte dell'opinione pubblica e della politica mondiale: anche se la ricerca nel settore sta aumentando a dismisura negli ultimi tempi, le risposte al fenomeno sono ancora insufficienti ed inadeguate. Ulteriori ritardi nella lotta ai cambiamenti climatici aumenteranno, di molto, i rischi sulla salute umana e pregiudicheranno la possibilità per i governi della Terra di garantire il diritto fondamentale della salute ai propri cittadini.

Fino a pochi decenni fa, "solamente" gli innumerevoli modelli matematici prevedevano che il clima del Pianeta stesse cambiando troppo rapidamente, mentre la stragrande maggioranza delle nazioni e pochissimi esponenti del mondo scientifico mostravano un completo scetticismo. Oggi si è di fronte a fenomeni climatici sempre più estremi, frequenti e devastanti e nessuno ha più dubbi in merito al fatto che siano in atto importanti mutazioni nel clima del Pianeta a causa, soprattutto, dell'operato umano. La comunità scientifica ha, infatti, da decenni additato contro le attività umane, indicandole come responsabili della crisi climatica, in particolare a causa dell’aumento dei gas serra immessi nell’atmosfera. La concentrazione di gas serra nell’atmosfera ha raggiunto livelli record: l’anidride carbonica è aumentata del 147%, il metano del 259% e il protossido di azoto del 123% rispetto ai livelli preindustriali (dati tratti dal [WWF Italia](https://www.wwf.it/cosa-facciamo/clima/cambiamenti-climatici/)).

**Deforestazione**

Tra le tante cause dell'operato umano c'è anche il problema della deforestazione che colpisce molte aree della Terra, in particolare quelle da cui è possibile ricavare legname pregiato e che sono anche quelle, purtroppo, più ricche di biodiversità come le foreste pluviali e tropicali. Negli ultimi 30 anni la superficie forestale a livello mondiale si è ridotta di oltre 420 milioni di ettari, con un ritmo, che dal 2010, è di circa 4,7 milioni di ettari all’anno. Il ritmo con cui l'uomo sta consumando le foreste del Pianeta è preoccupante, peraltro i dati sulla deforestazione non tengono conto delle aree che ogni anno vengono irrimediabilmente degradate a causa del sovrasfruttamento delle foreste, dell’inquinamento e delle conseguenze di tutti quei disastri, naturali o di origine antropica, originati da eventi climatici, incendi e da conflitti.

I dati raccolti dalla [World Bank](https://data.worldbank.org/indicator/AG.LND.FRST.ZS?end=2020&start=1990&view=chart) mostrano una significativa decrescita di area occupata dalle foreste sui territori dei paesi del Mondo, dal 1990 al 2016:

**Virus zoonotici e salto di specie**

Le malattie infettive emergenti minacciano in modo significativo la salute pubblica globale e la sicurezza socio-economica. La maggior parte dei focolai di malattie infettive è causata da virus zoonotici/trasmessi da vettori. I pipistrelli e i roditori sono i due più importanti "ospiti serbatoio" di molti virus zoonotici che possono attraversare le barriere delle specie per infettare anche gli esseri umani, mentre le zanzare e le zecche sono i principali vettori consolidati di molte malattie arbovirali.

I dati utilizzati sono raccolti dal database, recentemente aggiornato, di virus zoonotici  [ZOVER](http://www.mgc.ac.cn/ZOVER), in cui si incorporano due insiemi di dati recentemente rilasciati, DBatVir e DRodVir, rispettivamente per i virus associati ai pipistrelli e ai roditori, e si raccolgono ulteriori conoscenze aggiornate sui virus associati alle zanzare e alle zecche per creare una risorsa online completa per zoonosi e vettori virus trasmessi. Si riportano i dati relativi al numero di sequenze nucleotidiche associate ai virus trasmessi da pipistrelli, roditori, mosche e zecche, del genoma umano disponibili sulla banca dati [GenBank](https://www.ncbi.nlm.nih.gov/genbank/), il cui aumento esponenziale implica un continuo hotspot di virus zoonotici negli ultimi anni.

```python
df_virus = pd.read_csv('data/virus_sequences.csv')
df_virus
```

I dati raccolti dal database di virus zoonotici [ZOVER](http://www.mgc.ac.cn/ZOVER), non indicano alcuna incertezza  sul numero di sequenze nucleotidiche associate ai virus trasmessi da pipistrelli, roditori, mosche e zecche, sequenziati negli anni. Si può immaginare che negli ultimi decenni ci sia stato una maggiore "attenzione scientifica" che ha portato ad un numero maggiore, e probabilmente più vicino alla realtà, di virus sequenziati, rispetto al 1995. Per un'analisi dell'andamento generale dell'aumento del numero di sequenze nucleotidiche associate ai virus sequenziati, è stata considerata un'incertezza poissoniana (\$\sqrt{n}$, ove n sono i valori misurati per zanzare, zecche, roditori e pipistrelli negli anni).

Sulla base dei dati presentati nella tabella precendenta è possibile osservare un andamento interessante, anno per anno, del proliferare di nuove sequenze nucleotidiche e vettori di virus a partire da zanzare, zecche, roditori e pipistrelli:

```python
bat_err = np.sqrt(df_virus['Bat'])
mosquito_err = np.sqrt(df_virus['Mosquito'])
rodent_err=np.sqrt(df_virus['Rodent'])
tick_err=np.sqrt(df_virus['Tick'])
plt.figure(figsize=(18,11))
plt.scatter(df_virus['Year'], df_virus['Bat'], marker='o', label='sequenze nucleotidiche virali associate ai pipistrelli', s=10, color='blue')
plt.errorbar(df_virus['Year'], df_virus['Bat'], fmt='o', markersize=2, yerr=bat_err, label='errore')
plt.scatter(df_virus['Year'], df_virus['Mosquito'], marker='o', label='sequenze nucleotidiche virali associate ai vettori zanzare', s= 10, color='red')
plt.errorbar(df_virus['Year'], df_virus['Mosquito'], fmt='o', markersize=2, yerr=mosquito_err, label='errore')
plt.scatter(df_virus['Year'], df_virus['Rodent'], marker='o', label='sequenze nucleotidiche virali associate ai roditori', s= 10, color='orange')
plt.errorbar(df_virus['Year'], df_virus['Rodent'], fmt='o', markersize=2, yerr=rodent_err, label='errore')
plt.scatter(df_virus['Year'], df_virus['Tick'], marker='o',label='sequenze nucleotidiche virali associate ai vettori zecche', s= 10,  color='green')
plt.errorbar(df_virus['Year'], df_virus['Tick'], fmt='o', markersize=2, yerr=tick_err, label='errore')
plt.title('Numero di sequenze nucleotidiche e vettori di virus ogni anno nel Mondo')
plt.ylabel('sequenza relativa')
plt.xlabel('anno')
plt.legend()
plt.show()
```

```python
bat_err = np.sqrt(df_virus['Bat'])
mosquito_err = np.sqrt(df_virus['Mosquito'])
rodent_err=np.sqrt(df_virus['Rodent'])
tick_err=np.sqrt(df_virus['Tick'])
fig, ax = plt.subplots(2, 2, figsize=(18,11))
ax[0,0].errorbar(df_virus['Year'], df_virus['Bat'], fmt='o', markersize=2, yerr=bat_err, color='blue')
ax[0,0].set_title('sequenze nucleotidiche virali associate ai pipistrelli')
ax[0,1].errorbar(df_virus['Year'], df_virus['Mosquito'], fmt='o', markersize=2, yerr=mosquito_err, color='red')
ax[0,1].set_title('sequenze nucleotidiche virali associate ai vettori zanzare')
ax[1,0].errorbar(df_virus['Year'], df_virus['Rodent'], fmt='o', markersize=2, yerr=rodent_err, color='orange')
ax[1,0].set_title('sequenze nucleotidiche virali associate ai roditori')
ax[1,1].errorbar(df_virus['Year'], df_virus['Tick'], fmt='o', markersize=2, yerr=tick_err, color='green')
ax[1,1].set_title('sequenze nucleotidiche virali associate ai vettori zecche')
fig.suptitle('Nuove sequenze nucleotidiche e vettori di virus ogni anno nel Mondo')
```

L'aumento esponenziale di nuove sequenze nucleotidiche nel genoma umano derivante da virus zoonotici negli ultimi anni non è casuale; le foreste hanno un ruolo cruciale nella regolazione di batteri, virus e, di conseguenza, delle malattie. Delle foreste e degli ecosistemi in buono stato di salute limitano l'esposizione e l'impatto di agenti patogeni, comprese le zoonosi (malattie trasmissibili, direttamente o indirettamente, da animali all’uomo, come il COVID-19), attraverso un effetto diluente, o tampone, limitando così le possibilità di propagazione di virus. Al contrario, quando la biodiversità è sottoposta a pressioni che ne riducono lo stato di salute, il servizio eco-sistemico di controllo e regolazione di batteri, virus e delle malattie è fortemente compromesso ed è più probabile che emergano agenti patogeni.

E' possibile, quindi, supporre la presenza di un qualche correlazione tra i dati della percentuale di copertura forestale ogni anno nel Mondo e l'aumento di sequenze nucleotidiche e di vettori di virus sequenziati. Si badi bene che la deforestazione, rientra nella tematica molto più complessa dei cambiamenti climatici, per cui, per un'analisi approfondita, si dovrebbero tenere in considerazione anche i dati relativi all'aumento di C02 nell'atmosfera negli anni, all'aumento delle temperature globali negli anni, al tasso di inquinamento delle acque, ecc ecc; per tale motivi non è possibile andar oltre la semplice osservazione sperimentale di una qualche correlazione tra i vali della percentuale di copertura forestale ogni anno nel Mondo e l'aumento di sequenze nucleotidiche e di vettori di virus sequenziati. Non è possibile presentare una relazione di "causa-effetto".

Come ipotesi di lavoro è stata considerata una curva di fit esponenziale decrescente, come andamento approssimativo dei parametri nel tempo della percentuale di copertura forestale nel Mondo e del sequenziamento di nuovi virus trasmessi da pipistrelli, roditori, mosche e zecche.

```python
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
```

```python
popt, pcov = curve_fit(func, df_deforestation_perc['World forest area(%)'], (df_virus['Bat'] + np.sqrt(df_virus['Bat'])), sigma=np.sqrt(df_virus['Bat']), p0=[8E156, 11., 0], absolute_sigma=True, maxfev=5000)
print('Fit \n% di foreste vs sequenze nucleotidiche virali associate ai pipistrelli')
print("Parametri del Fit: ", popt)
print("Matrice di covarianza:", pcov)
print("a = {:.3f}+/-{:.3f}".format(popt[0],math.sqrt(pcov[0,0])))
print("b = {:.3f}+/-{:.3f}".format(popt[1],math.sqrt(pcov[1,1])))
print("c = {:.3f}+/-{:.3f}".format(popt[2],math.sqrt(pcov[2,2])))

plt.figure(figsize=(16,9))
plt.scatter(df_deforestation_perc['World forest area(%)'], (df_virus['Bat'] + np.sqrt(df_virus['Bat'])), marker='o', label='Percentuale di copertura forestale e aumento delle sequenze nucleotidiche virali associate ai pipistrelli', s=10, color='blue')
plt.plot(df_deforestation_perc['World forest area(%)'], func(df_deforestation_perc['World forest area(%)'], *popt), 'r-', label='Parametri del fit: a={:5.3f}, b={:5.3f}, c={:5.3f}'.format(*popt))
plt.title('Percentuale di copertura forestale ogni anno nel Mondo e aumento di sequenze nucleotidiche associate ai pipistrelli')
plt.ylabel('sequenze nucleotidiche e vettori di virus')
plt.xlabel('% di copertura forestale')
#plt.legend()
plt.show()
```

```python
popt, pcov = curve_fit(func, df_deforestation_perc['World forest area(%)'], (df_virus['Mosquito'] + np.sqrt(df_virus['Mosquito'])), sigma=np.sqrt(df_virus['Mosquito']), p0=[8E87, 6.299, 0], absolute_sigma=True, maxfev=5000)
print('Fit \n% di foreste vs sequenze nucleotidiche virali associate ai vettori zanzare')
print("Parametri del Fit: ", popt)
print("Matrice di covarianza:", pcov)
print("a = {:.3f}+/-{:.3f}".format(popt[0],math.sqrt(pcov[0,0])))
print("b = {:.3f}+/-{:.3f}".format(popt[1],math.sqrt(pcov[1,1])))
print("c = {:.3f}+/-{:.3f}".format(popt[2],math.sqrt(pcov[2,2])))

plt.figure(figsize=(16,9))
plt.scatter(df_deforestation_perc['World forest area(%)'], (df_virus['Bat'] + np.sqrt(df_virus['Bat'])), marker='o', label='Percentuale di copertura forestale e aumento delle sequenze nucleotidiche virali associate ai vettori zanzare', s=10, color='red')
plt.plot(df_deforestation_perc['World forest area(%)'], func(df_deforestation_perc['World forest area(%)'], *popt), 'r-', label='Parametri del fit: a={:5.3f}, b={:5.3f}, c={:5.3f}'.format(*popt))
plt.title('Percentuale di copertura forestale ogni anno nel Mondo e aumento di sequenze virali associate ai vettori zanzare')
plt.ylabel('sequenze nucleotidiche e vettori di virus')
plt.xlabel('% di copertura forestale')
#plt.legend()
plt.show()
```

```python
popt, pcov = curve_fit(func, df_deforestation_perc['World forest area(%)'], (df_virus['Rodent'] + np.sqrt(df_virus['Rodent'])), sigma=np.sqrt(df_virus['Rodent']), p0=[1E69, 4.884, 0], absolute_sigma=True, maxfev=5000)
print('Fit \n% di foreste vs sequenze nucleotidiche virali associate ai roditori')
print("Parametri del Fit: ", popt)
print("Matrice di covarianza:", pcov)
print("a = {:.3f}+/-{:.3f}".format(popt[0],math.sqrt(pcov[0,0])))
print("b = {:.3f}+/-{:.3f}".format(popt[1],math.sqrt(pcov[1,1])))
print("c = {:.3f}+/-{:.3f}".format(popt[2],math.sqrt(pcov[2,2])))

plt.figure(figsize=(16,9))
plt.scatter(df_deforestation_perc['World forest area(%)'], (df_virus['Rodent'] + np.sqrt(df_virus['Rodent'])), marker='o', label='Percentuale di copertura forestale e aumento delle sequenze nucleotidiche virali associate ai roditori', s=10, color='green')
plt.plot(df_deforestation_perc['World forest area(%)'], func(df_deforestation_perc['World forest area(%)'], *popt), 'r-', label='Parametri del fit: a={:5.3f}, b={:5.3f}, c={:5.3f}'.format(*popt))
plt.title('Percentuale di copertura forestale ogni anno nel Mondo e aumento di sequenze virali associate ai roditori')
plt.ylabel('sequenze nucleotidiche e vettori di virus')
plt.xlabel('% di copertura forestale')
#plt.legend()
plt.show()
```

```python
popt, pcov = curve_fit(func, df_deforestation_perc['World forest area(%)'], (df_virus['Tick'] + np.sqrt(df_virus['Tick'])), sigma=np.sqrt(df_virus['Tick']), p0=[4E118, 8.618, 0], absolute_sigma=True, maxfev=5000)
print('Fit \n% di foreste vs sequenze nucleotidiche virali associate ai vettori zecche')
print("Parametri del Fit: ", popt)
print("Matrice di covarianza:", pcov)
print("a = {:.3f}+/-{:.3f}".format(popt[0],math.sqrt(pcov[0,0])))
print("b = {:.3f}+/-{:.3f}".format(popt[1],math.sqrt(pcov[1,1])))
print("c = {:.3f}+/-{:.3f}".format(popt[2],math.sqrt(pcov[2,2])))

plt.figure(figsize=(16,9))
plt.scatter(df_deforestation_perc['World forest area(%)'], (df_virus['Tick'] + np.sqrt(df_virus['Tick'])), marker='o', label='Percentuale di copertura forestale e aumento delle sequenze nucleotidiche virali associate ai vettori zanzare', s=10, color='red')
plt.plot(df_deforestation_perc['World forest area(%)'], func(df_deforestation_perc['World forest area(%)'], *popt), 'r-', label='Parametri del fit: a={:5.3f}, b={:5.3f}, c={:5.3f}'.format(*popt))
plt.title('Percentuale di copertura forestale ogni anno nel Mondo e aumento di sequenze virali associate ai vettori zecche')
plt.ylabel('sequenze nucleotidiche e vettori di virus')
plt.xlabel('% di copertura forestale')
#plt.legend()
plt.show()
```

Le varie curve di fit esponenziale decrescenti rintracciate per ogni singolo caso analizzato, così definite, presentano dei parametri con poco senso fisico. Si consideri una riparametrizzazione della funzione esponenziale, a partire dalla considerazione della stessa ai limiti in cui: 
a) la percentuale di copertula forestale nel Mondo tenda allo 0%; 
b) la percentuale di copertula forestale nel Mondo tenda al 100%.
Si consideri la funzione esponenziale utilizzata per la realizzazione dei fit:

$\displaystyle f(x;a,b,c) = a\,e^{-bx}+c$

$\displaystyle f(x;a,b,c)$ al limite in cui la percentuale di copertula forestale nel Mondo tenda allo 0%, assumerà la forma:

$\displaystyle N_0 = f(0;a,b,c) = a+c$

Sotto detta ipotesi, ci si può aspettare un numero tendente all'infinito di sequenze nucleotidiche e vettori di virus.

$\displaystyle f(x;a,b,c)$ al limite in cui la percentuale di copertula forestale nel Mondo tenda al 100%, assumerà la forma:

$\displaystyle N_1 = f(100;a,b,c) = a\,e^{-b}+c$

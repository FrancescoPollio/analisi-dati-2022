# Cambiamenti climatici ed epidemie: nuove sfide per l'umanità
     
     
   *Il cambiamento climatico rappresenta la più grande minaccia del XXI secolo per la salute globale 
   (Organizzazione Mondiale della Sanità, 2003)*



I cambiamenti climatici rappresentano una delle maggiori sfide del XXI secolo e il loro impatto sulla Terra è una delle priorità per la tutela della salute pubblica; gli effetti, infatti, di detti cambiamenti, sempre più violenti ed estremi, minacciano la società contemporanea, sotto ogni aspetto e in tutti i suoi ambiti. I rischi sanitari, potenzialmente associati al clima che cambia, devono assolutamente ricevere maggiore attenzione da parte dell'opinione pubblica e della politica mondiale: anche se la ricerca nel settore sta aumentando a dismisura negli ultimi tempi, le risposte al fenomeno sono ancora insufficienti ed inadeguate. Ulteriori ritardi nella lotta ai cambiamenti climatici aumenteranno, di molto, i rischi sulla salute umana e pregiudicheranno la possibilità per i governi della Terra di garantire il diritto fondamentale della salute ai propri cittadini.

Fino a pochi decenni fa, "solamente" gli innumerevoli modelli matematici prevedevano che il clima del Pianeta stesse cambiando troppo rapidamente, mentre la stragrande maggioranza delle nazioni e pochissimi esponenti del mondo scientifico mostravano un completo scetticismo. Oggi si è di fronte a fenomeni climatici sempre più estremi, frequenti e devastanti e nessuno ha più dubbi in merito al fatto che siano in atto importanti mutazioni nel clima del Pianeta a causa, soprattutto, dell'operato umano. La comunità scientifica ha, infatti, da decenni additato contro le attività umane, indicandole come responsabili della crisi climatica, in particolare a causa dell’aumento dei gas serra immessi nell’atmosfera. La concentrazione di gas serra nell’atmosfera ha raggiunto livelli record: l’anidride carbonica è aumentata del 147%, il metano del 259% e il protossido di azoto del 123% rispetto ai livelli preindustriali (dati tratti dal [WWF Italia](https://www.wwf.it/cosa-facciamo/clima/cambiamenti-climatici/)).

**Deforestazione**

Tra le tante cause dell'operato umano c'è anche il problema della deforestazione che colpisce molte aree della Terra, in particolare quelle da cui è possibile ricavare legname pregiato e che sono anche quelle, purtroppo, più ricche di biodiversità come le foreste pluviali e tropicali. Negli ultimi 30 anni la superficie forestale a livello mondiale si è ridotta di oltre 420 milioni di ettari, con un ritmo, che dal 2010, è di circa 4,7 milioni di ettari all’anno. Il ritmo con cui l'uomo sta consumando le foreste del Pianeta è preoccupante, peraltro i dati sulla deforestazione non tengono conto delle aree che ogni anno vengono irrimediabilmente degradate a causa del sovrasfruttamento delle foreste, dell’inquinamento e delle conseguenze di tutti quei disastri, naturali o di origine antropica, originati da eventi climatici, incendi e da conflitti.

I dati raccolti dalla [World Bank](https://data.worldbank.org/indicator/AG.LND.FRST.ZS?end=2020&start=1990&view=chart) mostrano una significativa decrescita di area occupata dalle foreste sui territori dei paesi del Mondo, dal 1990 al 2016 (la tabella parte dal 1960 e giunge sino al 2020, ma è incompleta per il range che va dal 1960 al 1995 e dal 2017 ad oggi, sebbene l'ultimo aggiornamento della stessa è risalente al 5/5/22):

**Virus zoonotici e salto di specie**

Le malattie infettive emergenti minacciano in modo significativo la salute pubblica globale e la sicurezza socio-economica. La maggior parte dei focolai di malattie infettive è causata da virus zoonotici/trasmessi da vettori. I pipistrelli e i roditori sono i due più importanti "ospiti serbatoio" di molti virus zoonotici che possono attraversare le barriere delle specie per infettare anche gli esseri umani, mentre le zanzare e le zecche sono i principali vettori consolidati di molte malattie arbovirali.

I dati utilizzati sono raccolti dal database, recentemente aggiornato, di virus zoonotici  [ZOVER](http://www.mgc.ac.cn/ZOVER), in cui si incorporano due insiemi di dati recentemente rilasciati, DBatVir e DRodVir, rispettivamente per i virus associati ai pipistrelli e ai roditori, e si raccolgono ulteriori conoscenze aggiornate sui virus associati alle zanzare e alle zecche per creare una risorsa online completa per zoonosi e vettori virus trasmessi. Si riportano i dati relativi al numero di sequenze nucleotidiche associate ai virus trasmessi da pipistrelli, roditori, mosche e zecche, del genoma umano disponibili sulla banca dati [GenBank](https://www.ncbi.nlm.nih.gov/genbank/), il cui aumento esponenziale implica un continuo hotspot di virus zoonotici negli ultimi anni.

```python
df_virus = pd.read_csv('data/virus_sequences.csv')
df_virus
```

Sulla base dei dati presentati nella tabella precendenta è possibile osservare un andamento interessante, anno per anno, del proliferare di nuove sequenze nucleotidiche e vettori di virus a partire da zanzare, zecche, roditori e pipistrelli:

DA AGGIUSTARE

L'aumento esponenziale di nuove sequenze nucleotidiche nel genoma umano derivante da virus zoonotici negli ultimi anni non è casuale;

le foreste (e in genere tutti gli ecosistemi naturali e semi-naturali) siano preziose perché hanno un ruolo cruciale nella regolazione delle malattie. Le foreste e gli ecosistemi in buono stato di salute limitano l'esposizione e l'impatto di agenti patogeni, comprese le zoonosi (malattie trasmissibili, direttamente o indirettamente, da animali all’uomo, come la COVID-19), attraverso un effetto diluente o tampone, limitando così le possibilità di propagazione di agenti patogeni dalla fauna selvatica alle persone. Al contrario, quando la biodiversità è sottoposta a pressioni che ne riducono lo stato di salute, il servizio eco-sistemico di controllo e regolazione delle malattie è compromesso ed è più probabile che emergano agenti patogeni.

La distruzione, la frammentazione e la degradazione delle foreste o di altri habitat aprono spazi all' avvicinamento delle persone alla fauna selvatica, aumentando il rischio di trasmissione di malattie zoonotiche.


Il database ZOVER sarà una risorsa preziosa per virologi, zoologi ed epidemiologi per comprendere meglio la diversità e le dinamiche dei virus zoonotici e dei virus trasmessi da vettori e condurre una sorveglianza efficace per monitorare il potenziale spillover interspecie per una prevenzione e un controllo efficienti delle future malattie zoonotiche emergenti.

```python
plt.figure(figsize=(15,10))
plt.scatter(df_deforestation_perc['World forest area(%)'], df_virus['Bat'], marker='o', label='Percentuale di copertura forestale e aumento delle sequenze nucleotidiche virali associate ai pipistrelli', color='blue')
plt.errorbar(df_deforestation_perc['World forest area(%)'], df_virus['Bat'], fmt='o', markersize=2, xerr=def_noise, yerr=bat_noise, label='errore')
plt.scatter(df_deforestation_perc['World forest area(%)'], df_virus['Mosquito'], marker='o', label='Percentuale di copertura forestale e aumento delle sequenze nucleotidiche virali associate ai vettori zanzare', color='red')
plt.errorbar(df_deforestation_perc['World forest area(%)'], df_virus['Mosquito'], fmt='o', markersize=2, xerr=def_noise, yerr=mosquito_noise, label='errore')
plt.scatter(df_deforestation_perc['World forest area(%)'], df_virus['Rodent'], marker='o', label='Percentuale di copertura forestale e aumento delle sequenze nucleotidiche virali associate ai roditori', color='orange')
plt.errorbar(df_deforestation_perc['World forest area(%)'], df_virus['Rodent'], fmt='o', markersize=2, xerr=def_noise, yerr=rodent_noise, label='errore')
plt.scatter(df_deforestation_perc['World forest area(%)'], df_virus['Tick'], marker='o', label='Percentuale di copertura forestale e aumento delle sequenze nucleotidiche virali associate ai vettori zecche', color='green')
plt.errorbar(df_deforestation_perc['World forest area(%)'], df_virus['Tick'], fmt='o', markersize=2, xerr=def_noise, yerr=tick_noise, label='errore')
plt.title('Percentuale di copertura forestale ogni anno nel Mondo e aumento di sequenze nucleotidiche e di vettori di virus')
plt.ylabel('sequenze nucleotidiche e vettori di virus')
plt.xlabel('% di copertura forestale')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16,9))
ax[0,0].errorbar(df_deforestation_perc['World forest area(%)'], df_virus['Bat'], fmt='o', markersize=2, xerr=def_noise, yerr=bat_noise, color='blue')
ax[0,0].set_title('% di foreste e sequenze nucleotidiche virali associate ai pipistrelli')
ax[0,1].errorbar(df_deforestation_perc['World forest area(%)'], df_virus['Mosquito'], fmt='o', markersize=2, xerr=def_noise, yerr=mosquito_noise, color='red')
ax[0,1].set_title('% di foreste e sequenze nucleotidiche virali associate ai vettori zanzare')
ax[1,0].errorbar(df_deforestation_perc['World forest area(%)'], df_virus['Rodent'], fmt='o', markersize=2, xerr=def_noise, yerr=rodent_noise, color='orange')
ax[1,0].set_title('% di foreste e sequenze nucleotidiche virali associate ai roditori')
ax[1,1].errorbar(df_deforestation_perc['World forest area(%)'], df_virus['Tick'], fmt='o', markersize=2, xerr=def_noise, yerr=tick_noise, color='green')
ax[1,1].set_title('% di foreste e sequenze nucleotidiche virali associate ai vettori zecche')
fig.suptitle('Percentuale di copertura forestale ogni anno nel Mondo e aumento di sequenze nucleotidiche e di vettori di virus')
```

```python
def poly(x, a, b, c, d):
    return a + b * x + c * x ** 2 + d * x ** 3

popt, cov = curve_fit(poly, df_deforestation_perc['World forest area(%)'], df_virus['Bat'],  p0=[3.719 * 10 ** 7, -2.371 * 10 ** 6, 3.778 * 10 ** 4, 0])

print('Fit con limiti:')
print('Parametri del fit:', popt)
print('Matrice di covarianza:', cov)

plt.figure(figsize=(9,6), dpi=100)
plt.errorbar(df_deforestation_perc['World forest area(%)'], df_virus['Bat'], fmt='o', markersize=2, xerr=def_noise, yerr=bat_noise, label='errore')
plt.title('Percentuale di copertura forestale ogni anno nel Mondo e aumento di sequenze nucleotidiche e di vettori di virus')
plt.ylabel('sequenze nucleotidiche e vettori di virus')
plt.xlabel('% di copertura forestale')
plt.legend()

plt.plot(df_deforestation_perc['World forest area(%)'], poly(df_deforestation_perc['World forest area(%)'], *popt), 'r-', label='fit: a={:5.3f}, b={:5.3f}, c={:5.3f}, d={:5.3f}'.format(*popt))
```
